```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
current_centres = pd.read_excel('data/data.xlsx', sheet_name="Current Centres")
```

```python
current_centres.isna().sum()
```

```python
current_centres.isnull().sum()
```

```python
current_centres.dropna(inplace=True)
```

```python
current_centres.duplicated().sum()
```

```python
current_centres.drop_duplicates(inplace=True)
```

```python
current_centres.shape
```

```python
len(current_centres.columns)
```

```python
for column in current_centres.drop('CENTRE_NO', axis=1).columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=current_centres[column])
    plt.title(f'Box Plot of {column}')
    plt.xlabel(column)

plt.show()
```

```python
sns.histplot(current_centres, x="ANNUAL_REVENUE", bins=20)
plt.title("Annual Revenue Distribution of Current Centres")
plt.show()
```

```python
current_centres.columns
```

```python
encoding_map = {
    "A": 1, "B": 2, "C": 3, "D": 4, "E": 5, "F": 6
}
current_centres["AREA_AFFLUENCE_GRADE"] = current_centres["AREA_AFFLUENCE_GRADE"].map(encoding_map)
```

```python
sns.pairplot(current_centres, hue="ANNUAL_REVENUE")
plt.title("Pairplot of Current Centres")
plt.show()
```
